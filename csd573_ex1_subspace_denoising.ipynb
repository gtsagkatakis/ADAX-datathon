```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
num_samples = 1000
frequency = 5
x = np.linspace(0, 2 * np.pi, num_samples)
clean_signal_component_1=np.sin(frequency * x)
clean_signal_component_2=np.sin(frequency*2 * x + 5)
clean_signal = clean_signal_component_1 + clean_signal_component_2

plt.plot(x, clean_signal, label="Clean Signal")
plt.legend()
plt.title("Clean Sinusoidal Signal")
plt.show()
```

```python
noise_amplitude = 0.5
noisy_signal = clean_signal + noise_amplitude * np.random.randn(num_samples)

plt.plot(x, noisy_signal, label="Noisy Signal", alpha=0.7)
plt.plot(x, clean_signal, 'r', label="Clean Signal", linewidth=2)
plt.legend()
plt.title("Noisy Signal")
plt.show()
```

```python
# Normalize the clean signal to make it a unit vector
#clean_signal_normalized = clean_signal / np.linalg.norm(clean_signal)
clean_signal_normalized = clean_signal_component_1 / np.linalg.norm(clean_signal_component_1)


# Project the noisy signal onto the clean signal
projected_signal = np.dot(noisy_signal, clean_signal_normalized) * clean_signal_normalized

plt.plot(x, noisy_signal, label="Noisy Signal", alpha=0.7)
plt.plot(x, projected_signal, 'g', label="Projected/Denoised Signal", linewidth=2)
plt.legend()
plt.title("Projected Signal onto Clean Subspace")
plt.show()
```
